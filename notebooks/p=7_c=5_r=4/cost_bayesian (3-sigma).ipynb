```python
import sys
import pathlib
sys.path.append(str(pathlib.Path.cwd().parents[1]))

from control_chart.cost_bayesian import CostBayesianControlChart
from meter import Meter
from random_demand import stepwise_increasing

p = 7
c = 5
r = 4
meter = Meter(p, c, r)

for i in range(60):
    demand = stepwise_increasing(mu={0:1000, 13:1030, 26:1060, 39:1090}, sigma=10, n=52, random_state=i)
    control_chart = CostBayesianControlChart(mu0=1000, sigma=10, p=p, c=c, r=r, T=13, N=6, rule=3, p_=None)
    purchased = []
    for t, d in enumerate(demand):
        purchased.append(control_chart.purchase())
        control_chart.test(d, t)
    print(f"{i:>2} " , end="")
    results = meter(demand, purchased, verbose=True)
print("----------------------------------------------------------------------")
stats = meter.calculate_statistics()
```
